```python
input_folder = "/content"
output_csv = "/content/acousticbrainz_canciones.csv"

# Lista donde guardaremos todos los registros
rows = []

for filename in os.listdir(input_folder):
    if filename.endswith(".json"):
        file_path = os.path.join(input_folder, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            data = json.load(f)

        try:
            # Campos a extraer
            row = {
                "title": data["metadata"]["tags"].get("title", [None])[0] if "tags" in data["metadata"] else None,
                "artist_name": data["metadata"]["tags"].get("artist", [None])[0] if "tags" in data["metadata"] else None,
                "artist_mbid": data["metadata"]["tags"].get("musicbrainz_artistid", [None])[0] if "tags" in data["metadata"] else None,
                "track_mbid": data["metadata"]["tags"].get("musicbrainz_recordingid", [None])[0] if "tags" in data["metadata"] else None,
                "date": data["metadata"]["tags"].get("date", [None])[0] if "tags" in data["metadata"] else None,
                "length": data["metadata"]["audio_properties"].get("length", None) if "audio_properties" in data["metadata"] else None,
                "genre_dortmund": data["highlevel"]["genre_dortmund"]["value"],
                "genre_rosamerica": data["highlevel"]["genre_rosamerica"]["value"],
                "genre_tzanetakis": data["highlevel"]["genre_tzanetakis"]["value"],
                "sad": data["highlevel"]["mood_sad"]["all"].get("sad", None),
                "happy": data["highlevel"]["mood_happy"]["all"].get("happy", None),
                "party": data["highlevel"]["mood_party"]["all"].get("party", None),
                "relaxed": data["highlevel"]["mood_relaxed"]["all"].get("relaxed", None),
                "acoustic": data["highlevel"]["mood_acoustic"]["all"].get("acoustic", None),
                "danceable": data["highlevel"]["danceability"]["all"].get("danceable", None),
                "tonal": data["highlevel"]["tonal_atonal"]["all"].get("tonal", None),
                "bright": data["highlevel"]["timbre"]["all"].get("bright", None),
                "instrumental": data["highlevel"]["voice_instrumental"]["all"].get("instrumental", None),
            }

            rows.append(row)
        except Exception as e:
            print(f"Error procesando {filename}: {e}")

# Convertir a DataFrame y guardar en CSV
df = pd.DataFrame(rows)
df.to_csv(output_csv, index=False, encoding="utf-8-sig")
print(f"CSV generado: {output_csv}")
```

```python
import os
import json
import pandas as pd
from zipfile import ZipFile

# Subir el archivo ZIP (asumimos que el archivo subido se llama 'canc.zip')
from google.colab import files
#uploaded = files.upload()
zip_file_path = "39809542-d821-4150-98cd-d4814fe6d197-1.zip"

output_csv = "/content/acousticbrainz_canciones.csv"

# Lista donde guardaremos todos los registros
rows = []

try:
    with ZipFile(zip_file_path, 'r') as zf:
        # Itera sobre todos los nombres de archivo dentro del ZIP
        for filename in zf.namelist():
            # Filtra solo los archivos que terminan en .json
            if filename.endswith(".json"):
                print(f"Procesando archivo: {filename}")

                # zf.open(filename) reemplaza el open(file_path) con os.listdir
                with zf.open(filename, "r") as f:
                    # Lee el archivo JSON directamente desde el stream del ZIP
                    # Usamos .read() y .decode() porque zf.open en modo binario
                    # o texto puede requerir el contenido como string para json.loads
                    content = f.read().decode("utf-8")
                    data = json.loads(content)

                try:
                    row = {
                        "title": data["metadata"]["tags"].get("title", [None])[0] if "tags" in data["metadata"] else None,
                        "artist_name": data["metadata"]["tags"].get("artist", [None])[0] if "tags" in data["metadata"] else None,
                        "artist_mbid": data["metadata"]["tags"].get("musicbrainz_artistid", [None])[0] if "tags" in data["metadata"] else None,
                        "track_mbid": data["metadata"]["tags"].get("musicbrainz_recordingid", [None])[0] if "tags" in data["metadata"] else None,
                        "date": data["metadata"]["tags"].get("date", [None])[0] if "tags" in data["metadata"] else None,
                        "length": data["metadata"]["audio_properties"].get("length", None) if "audio_properties" in data["metadata"] else None,
                        "genre_dortmund": data["highlevel"]["genre_dortmund"]["value"],
                        "genre_rosamerica": data["highlevel"]["genre_rosamerica"]["value"],
                        "genre_tzanetakis": data["highlevel"]["genre_tzanetakis"]["value"],
                        "sad": data["highlevel"]["mood_sad"]["all"].get("sad", None),
                        "happy": data["highlevel"]["mood_happy"]["all"].get("happy", None),
                        "party": data["highlevel"]["mood_party"]["all"].get("party", None),
                        "relaxed": data["highlevel"]["mood_relaxed"]["all"].get("relaxed", None),
                        "acoustic": data["highlevel"]["mood_acoustic"]["all"].get("acoustic", None),
                        "danceable": data["highlevel"]["danceability"]["all"].get("danceable", None),
                        "tonal": data["highlevel"]["tonal_atonal"]["all"].get("tonal", None),
                        "bright": data["highlevel"]["timbre"]["all"].get("bright", None),
                        "instrumental": data["highlevel"]["voice_instrumental"]["all"].get("instrumental", None),
                    }
                    rows.append(row)
                except Exception as e:
                    print(f"Error procesando JSON de {filename}: {e}")

except FileNotFoundError:
    print(f"Error: No se encontró el archivo ZIP en '{zip_file_path}'. Asegúrate de que el nombre coincide con el archivo subido.")
except Exception as e:
    print(f"Error al abrir o procesar el archivo ZIP: {e}")


# Convertir a DataFrame y guardar en CSV
df = pd.DataFrame(rows)
df.to_csv(output_csv, index=False, encoding="utf-8-sig")
print(f"\n✅ CSV generado: {output_csv} con {len(df)} filas.")

# Descargar el csv en la pc
from google.colab import files
files.download(output_csv)
```
